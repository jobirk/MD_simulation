# 2D Ising model

The following properties are calculated in the following:
- The average energy for the $N^2\cdot1000$ steps equilibrium data $\left<H\right>$
- The average system magnetization $m = \left<S\right>$
- The specific heat at constant volume $C_V = \left(\frac{\partial U}{\partial T}\right)_h = \frac{\left<H^2\right> - \left<H\right>^2}{k_B T^2}$
- The magnetic susceptibility $\chi_T = \left(\frac{\partial m}{\partial h}\right)_h = \frac{\left<m^2\right> - \left<m\right>^2}{k_B T}$

These parameters are plotted against the temperature range.

```python
box = Ising_2D()
box.Ising_analysis(N=16, n_temp=50, steps_1=256000, steps_2=256000)
```

The mean energy (upper left plot) increases constantly with the temperature.

The slope of the energy, i.e. the specific heat (lower left plot), is increasing up to a critical temperature, the Curie temperature $T_C$ and then decreases again. The Curie Temperature is found at approximately $T_C \approx 0.3 K$.

The magnetisation (upper right plot) is either plus or minus one for temperatures below the Curie temperature. For higher energies, there is no magnetisation, i.e. the spins do not have a preferred direction and are orientated randomly which leads to a magnetisation of 0.

The magnetic susceptibility (lower left plot) is approximately zero for temperatures below the Curie temperature, then has a clear peak and decreases again, although it does not decrease to zero. This behaviour is reasonable as the magnetic susceptibility is the measure of how much the material is magnetised when an external magnetic field is applied. For temperatures slightly below the Curie temperature, where the magnetisation is found at values of for example $\pm 0.25$. Applying an external magnetic field would allow this state to transform into a state with a higher magnetisation (probably it would lead to $m=\pm1$). For temperatures above the Curie temperature this effect decreases since for these temperatures, the material transforms from a ferromagnetic material to a paramagnetic material. This can be seen in the way that the movement of the atoms inside the material is so high, that the spins of the electrons do not align corresponding to the magnetic field.

## Summary

Starting with randomly orientated spins, these spins have a preferred direction if the temperature is below a critical temperature, the "Curie temperature". In that case, both positive and negative magnetisation appears, depending on the initial structure.
If the temperature is above the curie temperature, no phase transition occurs, i.e. the spins change with the time but do not have a preferred orientation and thereby the magnetisation is approximately zero.

The simulation with the larger grid during the process of equilibration at a temperature below $T_c$ shows the phase transition quite nicely. However, here we can also see that in the case of a very large grid not all spins are aligned in the same direction as seen when using a smaller grid. Instead, there are some domains with spin +1 and others with spin -1. These are the so-called magnetic domains.
